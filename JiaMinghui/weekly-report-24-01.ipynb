## Predict Anything Model in Astronomy

训练一个跨模态的恒星参数预测模型，输入是`光谱、光变、参数`中的部分或者全部，并根据用户请求进行特定输出。


```python
import os
from tqdm import tqdm
from astropy.io import fits

root = '/data5/DataRelease/DR9_v2/fits/'
res = []
for dir, subdir, files in tqdm(os.walk(root)):
    for file in files:
        if file.endswith('.fits.gz'):
            try:
                with fits.open(os.path.join(dir, file)) as hdul:
                    header = hdul[0].header
                if header['NORM'] == 1:
                    res.append(dir + file)
            except Exception as e:
                print(e)

print(len(res))
with open('data.txt', 'w') as f:
    f.write('\n'.join(res))
```
